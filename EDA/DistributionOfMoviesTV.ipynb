```python
new_dataframe = data[['type', 'country']]

print(new_dataframe)
```

```python
filtered_data = data.dropna(subset=['country', 'type'])

# Group the data by 'country' and 'type' columns and get the counts
grouped_counts = filtered_data.groupby(['country', 'type']).size().reset_index(name='count')

# Separate the data into movies and TV shows
movies_data = grouped_counts[grouped_counts['type'] == 'Movie']
tv_shows_data = grouped_counts[grouped_counts['type'] == 'TV Show']

# Display the result for movies
print("Number of Movies per country:")
print(movies_data)

# Display the result for TV shows
print("\nNumber of TV Shows per country:")
print(tv_shows_data)
```

```python
movies_by_country = movies_data.groupby('country')['count'].sum().reset_index()
tv_shows_by_country = tv_shows_data.groupby('country')['count'].sum().reset_index()

# Merge the movies and TV shows data by country
combined_data = pd.merge(movies_by_country, tv_shows_by_country, on='country', suffixes=('_movies', '_tv_shows'))

# Calculate the total count of movies and TV shows
combined_data['total'] = combined_data['count_movies'] + combined_data['count_tv_shows']

# Get the top 5 countries based on the total count
top_countries = combined_data.nlargest(5, 'total')
data2 = top_countries.reset_index()
# Display the result

print(data2.drop('index',axis=1))
```

```python
top_countries_data = data[data['country'].isin(top_countries['country'])]

# Create the new DataFrame d3 with 'type' and 'country' columns
d3 = top_countries_data[['type', 'country','title']].copy()
# Display the result
print("DataFrame d3 with 'type' and 'country' columns from the Top 5 Countries:")
print(d3.reset_index(drop=True))
```

```python
labels = top_countries['country']
sizes = top_countries['total']
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'purple']
explode = (0.1, 0, 0, 0, 0)  # explode 1st slice

plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True, startangle=140)

plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Distribution of Movies and TV Shows in Top 5 Countries')
plt.show()
```
